```python
ratings = pd.read_csv("D:\\sliit\\3rd year 2nd sem\\IRW\\Mini project\\Book Recommender Sys\\data\\BX-Book-Ratings.csv", sep=";", on_bad_lines="skip", encoding="latin-1")
```

```python
ratings.rename(columns={
    "Book-Rating":"Rating"},inplace=True)
```

```python
ratings['User-ID'].value_counts()
```

```python
x = ratings['User-ID'].value_counts() > 200
```

```python
y = x[x].index
```

```python
ratings = ratings[ratings['User-ID'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books,on="ISBN")
```

```python
ratings_with_books.head(5)
```

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby('Title')['Rating'].count().reset_index()
```

```python
num_rating.head(5)
```

```python
num_rating.rename(columns={
    "Rating":"Num_of_rating"},inplace=True)
```

```python
num_rating.head(5)
```

```python
final_rating_table = ratings_with_books.merge(num_rating, on="Title")
```

```python
final_rating_table.head(2)
```

```python
final_rating_table.shape
```

```python
final_rating_table = final_rating_table[final_rating_table['Num_of_rating']>=50]
```

```python
final_rating_table.shape
```

```python
final_rating_table.head(2)
```

```python
final_rating_table.drop_duplicates(["User-ID","Title"], inplace=True)
```

```python
final_rating_table.shape
```

```python
book_pivot = final_rating_table.pivot_table(columns='User-ID',index='Title',values='Rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
#model building according to clustering algorithm nearest neighbors
#with brute technique
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
#algorithm returns 2 values as distance and suggestion
#these are the indexes of the output values
#testing if the model is fitted
#n_neighbors->how many recommendations
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
#testing if the recommendation system works well
def reccommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
b_name = 'Harry Potter and the Chamber of Secrets (Book 2)'
reccommend_book(b_name)
```
